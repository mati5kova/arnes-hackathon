```python
import geopandas as gpd
import matplotlib as plt
```

```python
gdf_poplave_zelo_redke = gpd.read_file("Data/DRSV_OPKP_ZR_POPL/DRSV_OPKP_ZR_POPL.shp")
gdf_poplave_redke = gpd.read_file("Data/DRSV_OPKP_REDKE_POPL/DRSV_OPKP_REDKE_POPL.shp")
gdf_poplave_pogoste = gpd.read_file("Data/DRSV_OPKP_POGOSTE_POPL/DRSV_OPKP_POGOSTE_POPL.shp")
gdf_poplave_razliv = gpd.read_file("Data/DRSV_OPKP_FUNKC_RAZLIV_POV/DRSV_OPKP_FUNKC_RAZLIV_POV.shp")
gdf_poplave_obmocja = gpd.read_file("Data/DRSV_POPDOG/DRSV_POPDOG_OBM.shp")
```

```python
poplave_list = [gdf_poplave_zelo_redke, gdf_poplave_redke, gdf_poplave_pogoste, gdf_poplave_razliv, gdf_poplave_obmocja]
```

```python
for poplave in poplave_list:
    print(poplave.shape)
    print(poplave.dtypes)
    print(poplave.head(3))
    print(poplave.crs)
    print(poplave.geometry.type.unique())

    poplave.plot(figsize=(10, 8))
    
```
